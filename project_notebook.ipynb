```python
def load_raw_data() -> pd.DataFrame:
    nsf_grants_data = pd.read_csv("all_grants_with_termination.csv")
    return nsf_grants_data
```

```python
#DELETE UNNECESSARY COLUMNS

def delete_unnecessary_columns(df: pd.DataFrame) -> pd.DataFrame:
    
    nsf_data_simplified = df.copy()

    cols_to_drop = [
        "agcy_id",
        "cfda_num",
        "org_code",
        "po_phone",
        "po_email",
        "po_sign_block_name",
        "awd_min_amd_letter_date",
        "awd_max_amd_letter_date",
        "awd_arra_amount",
        "awd_agcy_code",
        "fund_agcy_code",
        "pi_first_name",
        "pi_last_name",
        "pi_mid_init",
        "pi_full_name",
        "pi_email_addr",
        "nsf_id",
        "num_co_pis",
        "inst_street_address",
        "inst_country_name",
        "cong_dist_code",
        "st_cong_dist_code",
        "org_uei_num",
        "perf_st_code",
        "perf_zip_code",
        "perf_cong_dist",
        "perf_st_cong_dist",
        "perf_ctry_name",
        "pgm_ref_code",
        "pgm_ref_txt",
        "fund_oblg_fiscal_yr",
    ]

    nsf_data_simplified = nsf_data_simplified.drop(columns=[col for col in cols_to_drop if col in nsf_data_simplified.columns])
    return nsf_data_simplified
```

```python
def save_processed_data(df: pd.DataFrame):
    """
    Save the processed dataframe to CSV file.

    Args:
        df: Processed dataframe
    """
 
    output_path = "nsf_data_clean.csv"
    df.to_csv(output_path, index=False)
    print(f"Processed data saved to {output_path}")
```

```python
def main_processing_pipeline() -> pd.DataFrame:
    """
    Main processing pipeline that combines all cleaning steps.

    Returns:
        Fully processed dataframe ready for visualization
    """
    print("Loading raw data...")
    nsf_data = load_raw_data()
    print(nsf_data.head())

    print("Cleaning NSF data...")
    nsf_clean = delete_unnecessary_columns(nsf_data)
    print(nsf_clean.head())
    
    print("Saving processed data...")
    save_processed_data(nsf_clean)
    
    return nsf_clean
```

```python
processed_data = main_processing_pipeline()
print("Data processing complete!")
processed_data.head()
```

### Interactive Bar Chart: Grants by Directorate

This visualization shows the distribution of grants across NSF directorates with:
- **Year selector**: Choose "All years" for aggregated data or select a specific year
- **Metric selector**: Toggle between total grants and terminated grants
- **Interactive hover**: Highlights the selected directorate with detailed information

```python
# Prepare data aggregated by directorate and year
directorate_data = df.groupby(['dir_abbr', 'year']).agg({
    'awd_id': 'count',
    'terminated': lambda x: x.sum()  # Count terminated grants (terminated=True means 1)
}).reset_index()

directorate_data.columns = ['directorate', 'year', 'num_grants', 'terminated_grants']

# Convert year to string for consistency
directorate_data['year'] = directorate_data['year'].astype(str)

# Also prepare data for "All years"
directorate_all_years = df.groupby('dir_abbr').agg({
    'awd_id': 'count',
    'terminated': lambda x: x.sum()
}).reset_index()

directorate_all_years.columns = ['directorate', 'num_grants', 'terminated_grants']
directorate_all_years['year'] = 'All years'

# Combine both datasets
directorate_combined = pd.concat([
    directorate_data,
    directorate_all_years
], ignore_index=True)

print(directorate_combined.head(10))
print(f"\nYears available: {directorate_combined['year'].unique()}")
```

```python
# Create year dropdown selector
year_options = ['All years', '2020', '2021', '2022', '2023', '2024', '2025']

year_dropdown = alt.binding_select(
    options=year_options,
    name='Year: '
)
year_selection = alt.param(
    name='selected_year',
    value='All years',
    bind=year_dropdown
)

# Create metric selector (radio buttons)
metric_radio = alt.binding_radio(
    options=['num_grants', 'terminated_grants'],
    labels=['Total Grants', 'Terminated Grants'],
    name='Metric: '
)
metric_selection = alt.param(
    name='selected_metric',
    value='num_grants',
    bind=metric_radio
)

# Create hover selection for highlighting
hover = alt.selection_point(
    on='mouseover',
    fields=['directorate'],
    empty=False
)

# Create the bar chart
bar_chart = alt.Chart(directorate_combined).mark_bar().encode(
    x=alt.X(
        'directorate:N',
        sort='-y',
        title='Directorate',
        axis=alt.Axis(labelAngle=-45)
    ),
    y=alt.Y(
        'value:Q',
        title='Number of Grants'
    ),
    color=alt.condition(
        hover,
        alt.Color('directorate:N', legend=None, scale=alt.Scale(scheme='category20')),
        alt.value('lightgray')
    ),
    opacity=alt.condition(
        hover,
        alt.value(1.0),
        alt.value(0.7)
    ),
    tooltip=[
        alt.Tooltip('directorate:N', title='Directorate'),
        alt.Tooltip('year:N', title='Year'),
        alt.Tooltip('num_grants:Q', title='Total Grants', format=','),
        alt.Tooltip('terminated_grants:Q', title='Terminated Grants', format=',')
    ]
).transform_filter(
    alt.datum.year == year_selection
).transform_fold(
    ['num_grants', 'terminated_grants'],
    as_=['metric', 'value']
).transform_filter(
    alt.datum.metric == metric_selection
).properties(
    width=700,
    height=400,
    title='NSF Grants Distribution by Directorate'
).add_params(
    year_selection,
    metric_selection,
    hover
)

bar_chart
```

### Normalized Stacked Bar Chart: Termination Rates by Directorate

This visualization shows whether grant terminations are concentrated in specific directorates:
- **Normalized stacked bars (100%)**: Each bar shows the percentage of terminated vs. active grants per directorate
- **Year range filter**: Use the slider to filter data by year range
- **Interactive click**: Click on a directorate to see detailed statistics

```python
# Prepare data for termination analysis
# Create a column for termination status
df_termination = df.copy()
df_termination['status'] = df_termination['terminated'].apply(lambda x: 'Terminated' if x else 'Active')

# Group by directorate, year, and status
termination_by_dir = df_termination.groupby(['dir_abbr', 'year', 'status']).size().reset_index(name='count')
termination_by_dir.columns = ['directorate', 'year', 'status', 'count']

print(termination_by_dir.head(20))
print(f"\nDirectorates: {termination_by_dir['directorate'].unique()}")
print(f"\nYears: {sorted(termination_by_dir['year'].unique())}")
```

```python
# Create year range slider
year_range_slider = alt.binding_range(
    min=2020,
    max=2025,
    step=1,
    name='Year Range Start: '
)
year_range_start = alt.param(
    name='year_start',
    value=2020,
    bind=year_range_slider
)

year_range_slider_end = alt.binding_range(
    min=2020,
    max=2025,
    step=1,
    name='Year Range End: '
)
year_range_end = alt.param(
    name='year_end',
    value=2025,
    bind=year_range_slider_end
)

# Base chart with filtering
base = alt.Chart(termination_by_dir).transform_filter(
    (alt.datum.year >= year_range_start) & (alt.datum.year <= year_range_end)
).transform_aggregate(
    count='sum(count)',
    groupby=['directorate', 'status']
).transform_calculate(
    order="datum.status == 'Terminated' ? 0 : 1"
)


# Create main stacked bar chart (normalized percentage view)
stacked_chart = base.transform_joinaggregate(
    total='sum(count)',
    groupby=['directorate']
).transform_calculate(
    percentage='datum.count / datum.total * 100'
).mark_bar().encode(
    x=alt.X('directorate:N', title='Directorate', sort='-y', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('sum(count):Q', title='Number of Grants', stack='normalize'),
    color=alt.Color('status:N', 
                    scale=alt.Scale(domain=['Active', 'Terminated'], 
                                  range=['#2ecc71', '#e74c3c']),
                    title='Grant Status'),
    order=alt.Order('order:Q'),
    tooltip=[
        alt.Tooltip('directorate:N', title='Directorate'),
        alt.Tooltip('status:N', title='Status'),
        alt.Tooltip('count:Q', title='Number of Grants', format=','),
        alt.Tooltip('percentage:Q', title='Percentage', format='.2f')
    ]
).properties(
    width=700,
    height=400,
    title='Grant Termination Rates by Directorate (Normalized to 100%)'
).add_params(
    year_range_start,
    year_range_end
)

# Combine main chart with detail panel
final_chart = (stacked_chart).resolve_scale(
    color='shared'
)

final_chart
```

### Evolution of Terminated Grants Over Time by Directorate

This interactive line chart shows how terminated grants evolved over the years for each directorate:
- **Multi-line view**: Each line represents a directorate
- **Interactive hover**: Hover over the chart to see exact values for all directorates at that year
- **Details on demand**: Year ruler and values appear when hovering

```python
# Prepare data for line chart - terminated grants by directorate and year
terminated_evolution = df.groupby(['dir_abbr', 'year']).agg({
    'terminated': lambda x: x.sum()
}).reset_index()
terminated_evolution.columns = ['directorate', 'year', 'terminated_grants']

# Filter out directorates with very few grants for cleaner visualization
main_directorates = ['MPS', 'CSE', 'ENG', 'GEO', 'EDU', 'BIO', 'TIP', 'SBE']
terminated_evolution_filtered = terminated_evolution[terminated_evolution['directorate'].isin(main_directorates)]

print(terminated_evolution_filtered.head(20))
print(f"\nDirectorates included: {sorted(terminated_evolution_filtered['directorate'].unique())}")
```

```python
# Create a selection that chooses the nearest point based on year
nearest = alt.selection_point(
    nearest=True, 
    on='mouseover',
    fields=['year'], 
    empty=False
)

# The basic line chart
line = alt.Chart(terminated_evolution_filtered).mark_line(
    interpolate='monotone',
    strokeWidth=2
).encode(
    x=alt.X('year:Q', title='Year', axis=alt.Axis(format='d')),
    y=alt.Y('terminated_grants:Q', title='Number of Terminated Grants'),
    color=alt.Color('directorate:N', 
                   title='Directorate',
                   scale=alt.Scale(scheme='category10'))
)

# Transparent selectors across the chart
selectors = alt.Chart(terminated_evolution_filtered).mark_point().encode(
    x='year:Q',
    opacity=alt.value(0),
).add_params(nearest)

# Draw points on the lines, and highlight based on selection
points = line.mark_point(size=80).encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

# Draw text labels near the points, and highlight based on selection
text = line.mark_text(
    align='left', 
    dx=10, 
    dy=0,
    fontSize=11
).encode(
    text=alt.condition(nearest, 'terminated_grants:Q', alt.value(' '))
)

# Draw a vertical rule at the location of the selection
rules = alt.Chart(terminated_evolution_filtered).mark_rule(
    color='gray',
    strokeWidth=1
).encode(
    x='year:Q',
).transform_filter(nearest)

# Combine all layers
evolution_chart = alt.layer(
    line, selectors, points, rules, text
).properties(
    width=700,
    height=400,
    title='Evolution of Terminated Grants by Directorate (2020-2025)'
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_legend(
    labelFontSize=12,
    titleFontSize=13
)

evolution_chart
```
