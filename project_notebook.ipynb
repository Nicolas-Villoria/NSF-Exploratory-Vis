```python
def load_raw_data() -> pd.DataFrame:
    nsf_grants_data = pd.read_csv("all_grants_with_termination.csv")
    return nsf_grants_data
```

```python
#DELETE UNNECESSARY COLUMNS

def delete_unnecessary_columns(df: pd.DataFrame) -> pd.DataFrame:
    
    nsf_data_simplified = df.copy()

    cols_to_drop = [
        "agcy_id",
        "cfda_num",
        "org_code",
        "po_phone",
        "po_email",
        "po_sign_block_name",
        "awd_min_amd_letter_date",
        "awd_max_amd_letter_date",
        "awd_arra_amount",
        "awd_agcy_code",
        "fund_agcy_code",
        "pi_first_name",
        "pi_last_name",
        "pi_mid_init",
        "pi_full_name",
        "pi_email_addr",
        "nsf_id",
        "num_co_pis",
        "inst_street_address",
        "inst_country_name",
        "cong_dist_code",
        "st_cong_dist_code",
        "org_uei_num",
        "perf_st_code",
        "perf_zip_code",
        "perf_cong_dist",
        "perf_st_cong_dist",
        "perf_ctry_name",
        "pgm_ref_code",
        "pgm_ref_txt",
        "fund_oblg_fiscal_yr",
    ]

    nsf_data_simplified = nsf_data_simplified.drop(columns=[col for col in cols_to_drop if col in nsf_data_simplified.columns])
    return nsf_data_simplified
```

```python
def save_processed_data(df: pd.DataFrame):
    """
    Save the processed dataframe to CSV file.

    Args:
        df: Processed dataframe
    """
 
    output_path = "nsf_data_clean.csv"
    df.to_csv(output_path, index=False)
    print(f"Processed data saved to {output_path}")
```

```python
def main_processing_pipeline() -> pd.DataFrame:
    """
    Main processing pipeline that combines all cleaning steps.

    Returns:
        Fully processed dataframe ready for visualization
    """
    print("Loading raw data...")
    nsf_data = load_raw_data()
    print(nsf_data.head())

    print("Cleaning NSF data...")
    nsf_clean = delete_unnecessary_columns(nsf_data)
    print(nsf_clean.head())
    
    print("Saving processed data...")
    save_processed_data(nsf_clean)
    
    return nsf_clean
```

```python
processed_data = main_processing_pipeline()
print("Data processing complete!")
processed_data.head()
```

### Interactive Choropleth Map: Grants Distribution by State

This visualization shows how NSF grants are geographically distributed across U.S. states:
- **Year slider**: Select different years to see how the distribution of active grants changes over time
- **Color intensity**: Darker colors indicate more active grants
- **Interactive tooltips**: Hover over states to see detailed grant statistics

```python
from vega_datasets import data
import numpy as np

# First, ensure date columns are in datetime format
df['awd_eff_date'] = pd.to_datetime(df['awd_eff_date'], errors='coerce')
df['awd_exp_date'] = pd.to_datetime(df['awd_exp_date'], errors='coerce')

# Create a list to store results
results = []

# For each state and each year from 2020 to 2025
for state in df['inst_state_code'].unique():
    state_df = df[df['inst_state_code'] == state]
    
    for year in range(2020, 2026):
        # Create date range for the year
        year_start = pd.Timestamp(f'{year}-01-01')
        year_end = pd.Timestamp(f'{year}-12-31')
        
        # Count active grants: those where awd_eff_date <= year_end AND awd_exp_date >= year_start
        active_mask = (
            (state_df['awd_eff_date'] <= year_end) & 
            (state_df['awd_exp_date'] >= year_start)
        )
        num_active = active_mask.sum()
        
        # Count terminated grants: only show in 2025 (the year they were cancelled)
        if year == 2025:
            terminated_count = state_df['terminated'].sum()
        else:
            terminated_count = 0
        
        # Calculate percentage
        terminated_pct = (terminated_count / num_active * 100) if num_active > 0 else 0
        
        results.append({
            'state': state,
            'year': year,
            'num_grants': num_active,
            'terminated_grants': terminated_count,
            'terminated_pct': round(terminated_pct, 2)
        })

# Create DataFrame from results
grants_by_state_year = pd.DataFrame(results)

# Load US states geographic data
us_states = alt.topo_feature(data.us_10m.url, 'states')

# Create state FIPS mapping
state_fips = {
    'AL': 1, 'AK': 2, 'AZ': 4, 'AR': 5, 'CA': 6, 'CO': 8, 'CT': 9, 'DE': 10,
    'FL': 12, 'GA': 13, 'HI': 15, 'ID': 16, 'IL': 17, 'IN': 18, 'IA': 19,
    'KS': 20, 'KY': 21, 'LA': 22, 'ME': 23, 'MD': 24, 'MA': 25, 'MI': 26,
    'MN': 27, 'MS': 28, 'MO': 29, 'MT': 30, 'NE': 31, 'NV': 32, 'NH': 33,
    'NJ': 34, 'NM': 35, 'NY': 36, 'NC': 37, 'ND': 38, 'OH': 39, 'OK': 40,
    'OR': 41, 'PA': 42, 'RI': 44, 'SC': 45, 'SD': 46, 'TN': 47, 'TX': 48,
    'UT': 49, 'VT': 50, 'VA': 51, 'WA': 53, 'WV': 54, 'WI': 55, 'WY': 56,
    'DC': 11, 'PR': 72
}

# Add FIPS id to the dataframe
grants_by_state_year['id'] = grants_by_state_year['state'].map(state_fips)

# Remove rows without valid FIPS codes
grants_by_state_year = grants_by_state_year.dropna(subset=['id'])
grants_by_state_year['id'] = grants_by_state_year['id'].astype(int)

print(f"Data prepared for {len(grants_by_state_year)} state-year combinations")
print(f"\nYears available: {sorted(grants_by_state_year['year'].unique())}")
print(f"\nSample data for 2020 and 2025:")
print(grants_by_state_year[grants_by_state_year['year'].isin([2020, 2025])].head(10))
```

# Q2: How are the grants distributed per directorates? And for a certain year?

### Interactive Bar Chart: Grants by Directorate

This visualization shows the distribution of grants across NSF directorates with:
- **Year selector**: Choose "All years" for aggregated data or select a specific year
- **Metric selector**: Toggle between total grants and terminated grants
- **Interactive hover**: Highlights the selected directorate with detailed information

```python
# Create a list to store results for directorates
directorate_results = []

# For each directorate and each year from 2020 to 2025
for directorate in df['dir_abbr'].unique():
    dir_df = df[df['dir_abbr'] == directorate]
    
    for year in range(2020, 2026):
        # Create date range for the year
        year_start = pd.Timestamp(f'{year}-01-01')
        year_end = pd.Timestamp(f'{year}-12-31')
        
        # Count active grants: those where awd_eff_date <= year_end AND awd_exp_date >= year_start
        active_mask = (
            (dir_df['awd_eff_date'] <= year_end) & 
            (dir_df['awd_exp_date'] >= year_start)
        )
        num_active = active_mask.sum()
        
        # Count terminated grants: only show in 2025 (the year they were cancelled)
        if year == 2025:
            terminated_count = dir_df['terminated'].sum()
        else:
            terminated_count = 0
        
        directorate_results.append({
            'directorate': directorate,
            'year': str(year),
            'num_grants': num_active,
            'terminated_grants': terminated_count
        })

# Create DataFrame from results
directorate_data = pd.DataFrame(directorate_results)

# Also prepare data for "All years" - sum all active grants across all years
directorate_all_years = []
for directorate in df['dir_abbr'].unique():
    dir_df = df[df['dir_abbr'] == directorate]
    
    # Count total unique grants (using awd_id)
    total_grants = len(dir_df)
    terminated_total = dir_df['terminated'].sum()
    
    directorate_all_years.append({
        'directorate': directorate,
        'year': 'All years',
        'num_grants': total_grants,
        'terminated_grants': terminated_total
    })

directorate_all_years_df = pd.DataFrame(directorate_all_years)

# Combine both datasets
directorate_combined = pd.concat([
    directorate_data,
    directorate_all_years_df
], ignore_index=True)

print(directorate_combined.head(10))
print(f"\nYears available: {directorate_combined['year'].unique()}")
print(f"\nSample data for 2020 and 2025:")
print(directorate_combined[directorate_combined['year'].isin(['2020', '2025'])].head(20))
```

```python
# Create year dropdown selector
year_options = ['All years', '2020', '2021', '2022', '2023', '2024', '2025']

year_dropdown = alt.binding_select(
    options=year_options,
    name='Year: '
)
year_selection = alt.param(
    name='selected_year',
    value='All years',
    bind=year_dropdown
)

# Create hover selection for highlighting
hover = alt.selection_point(
    on='mouseover',
    fields=['directorate'],
    empty=False
)

# Create the bar chart
bar_chart = alt.Chart(directorate_combined).mark_bar().encode(
    x=alt.X(
        'directorate:N',
        sort='-y',
        title='Directorate',
        axis=alt.Axis(labelAngle=-45)
    ),
    y=alt.Y(
        'value:Q',
        title='Number of Grants'
    ),
    color=alt.condition(
        hover,
        alt.Color('directorate:N', legend=None, scale=alt.Scale(scheme='category20')),
        alt.value('lightgray')
    ),
    opacity=alt.condition(
        hover,
        alt.value(1.0),
        alt.value(0.7)
    ),
    tooltip=[
        alt.Tooltip('directorate:N', title='Directorate'),
        alt.Tooltip('year:N', title='Year'),
        alt.Tooltip('num_grants:Q', title='Total Grants', format=','),
        alt.Tooltip('terminated_grants:Q', title='Terminated Grants', format=',')
    ]
).transform_filter(
    alt.datum.year == year_selection
).transform_fold(
    ['num_grants', 'terminated_grants'],
    as_=['metric', 'value']
).properties(
    width=700,
    height=400,
    title='NSF Grants Distribution by Directorate'
).add_params(
    year_selection,
    hover
)

bar_chart
```

### Evolution of Active Grants Over Time by Directorate

- **Multi-line view**: Each line represents a directorate
- **Interactive hover**: Hover over the chart to see exact values for all directorates at that year
- **Details on demand**: Year ruler and values appear when hovering

```python
# Prepare data for line chart - reusing directorate_data
active_grants_df = directorate_data

# Filter to main directorates for cleaner visualization
main_directorates = ['MPS', 'CSE', 'ENG', 'GEO', 'EDU', 'BIO', 'TIP', 'SBE', 'O/D']
active_grants_filtered = active_grants_df[active_grants_df['directorate'].isin(main_directorates)]

print(active_grants_filtered.head(20))
print(f"\nDirectorates included: {sorted(active_grants_filtered['directorate'].unique())}")
```

```python
# Create a selection that chooses the nearest point based on year
nearest = alt.selection_point(
    nearest=True, 
    on='mouseover',
    fields=['year'], 
    empty=False
)

# The basic line chart
line = alt.Chart(active_grants_filtered).mark_line(
    interpolate='monotone',
    strokeWidth=2
).encode(
    x=alt.X('year:Q', title='Year', axis=alt.Axis(format='d')),
    y=alt.Y('num_grants:Q', title='Number of Active Grants'),
    color=alt.Color('directorate:N', 
                   title='Directorate',
                   scale=alt.Scale(scheme='category10'))
)

# Transparent selectors across the chart
selectors = alt.Chart(active_grants_filtered).mark_point().encode(
    x='year:Q',
    opacity=alt.value(0),
).add_params(nearest)

# Draw points on the lines, and highlight based on selection
points = line.mark_point(size=80).encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

# Draw text labels near the points, and highlight based on selection
text = line.mark_text(
    align='left', 
    dx=10, 
    dy=0,
    fontSize=11
).encode(
    text=alt.condition(nearest, 'num_grants:Q', alt.value(' '))
)

# Draw a vertical rule at the location of the selection
rules = alt.Chart(active_grants_filtered).mark_rule(
    color='gray',
    strokeWidth=1
).encode(
    x='year:Q',
).transform_filter(nearest)

# Combine all layers
active_grants_chart = alt.layer(
    line, selectors, points, rules, text
).properties(
    width=700,
    height=400,
    title='Evolution of Active Grants by Directorate (2020-2025)'
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_legend(
    labelFontSize=12,
    titleFontSize=13
)

active_grants_chart
```

```python
# Prepare data for termination impact analysis
termination_impact = []

# For each directorate, calculate termination metrics for 2025
for directorate in df['dir_abbr'].unique():
    dir_df = df[df['dir_abbr'] == directorate]
    
    # Count active grants in 2025
    year_start = pd.Timestamp('2025-01-01')
    year_end = pd.Timestamp('2025-12-31')
    active_mask = (
        (dir_df['awd_eff_date'] <= year_end) & 
        (dir_df['awd_exp_date'] >= year_start)
    )
    num_active_2025 = active_mask.sum()
    
    # Count terminated grants
    num_terminated = dir_df['terminated'].sum()
    
    # Calculate termination percentage
    termination_pct = (num_terminated / num_active_2025 * 100) if num_active_2025 > 0 else 0
    
    termination_impact.append({
        'directorate': directorate,
        'active_grants_2025': num_active_2025,
        'terminated_grants': num_terminated,
        'termination_pct': round(termination_pct, 2)
    })

# Create DataFrame
termination_impact_df = pd.DataFrame(termination_impact)

# Filter to main directorates and sort by terminated grants
main_directorates = ['MPS', 'CSE', 'ENG', 'GEO', 'EDU', 'BIO', 'TIP', 'SBE']
termination_impact_filtered = termination_impact_df[termination_impact_df['directorate'].isin(main_directorates)]
termination_impact_filtered = termination_impact_filtered.sort_values('terminated_grants', ascending=False)

print(termination_impact_filtered)
print(f"\nTotal terminated grants: {termination_impact_filtered['terminated_grants'].sum()}")
print(f"Average termination percentage: {termination_impact_filtered['termination_pct'].mean():.2f}%")
```
