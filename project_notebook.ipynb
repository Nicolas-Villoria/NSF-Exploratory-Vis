```python
def load_raw_data() -> pd.DataFrame:
    nsf_grants_data = pd.read_csv("all_grants_with_termination.csv")
    return nsf_grants_data
```

```python
#DELETE UNNECESSARY COLUMNS

def delete_unnecessary_columns(df: pd.DataFrame) -> pd.DataFrame:
    
    nsf_data_simplified = df.copy()

    cols_to_drop = [
        "agcy_id",
        "cfda_num",
        "org_code",
        "po_phone",
        "po_email",
        "po_sign_block_name",
        "awd_min_amd_letter_date",
        "awd_max_amd_letter_date",
        "awd_arra_amount",
        "awd_agcy_code",
        "fund_agcy_code",
        "pi_first_name",
        "pi_last_name",
        "pi_mid_init",
        "pi_full_name",
        "pi_email_addr",
        "nsf_id",
        "num_co_pis",
        "inst_street_address",
        "inst_country_name",
        "cong_dist_code",
        "st_cong_dist_code",
        "org_uei_num",
        "perf_st_code",
        "perf_zip_code",
        "perf_cong_dist",
        "perf_st_cong_dist",
        "perf_ctry_name",
        "pgm_ref_code",
        "pgm_ref_txt",
        "fund_oblg_fiscal_yr",
    ]

    nsf_data_simplified = nsf_data_simplified.drop(columns=[col for col in cols_to_drop if col in nsf_data_simplified.columns])
    return nsf_data_simplified
```

```python
def save_processed_data(df: pd.DataFrame):
    """
    Save the processed dataframe to CSV file.

    Args:
        df: Processed dataframe
    """
 
    output_path = "nsf_data_clean.csv"
    df.to_csv(output_path, index=False)
    print(f"Processed data saved to {output_path}")
```

```python
def main_processing_pipeline() -> pd.DataFrame:
    """
    Main processing pipeline that combines all cleaning steps.

    Returns:
        Fully processed dataframe ready for visualization
    """
    print("Loading raw data...")
    nsf_data = load_raw_data()
    print(nsf_data.head())

    print("Cleaning NSF data...")
    nsf_clean = delete_unnecessary_columns(nsf_data)
    print(nsf_clean.head())
    
    print("Saving processed data...")
    save_processed_data(nsf_clean)
    
    return nsf_clean
```

```python
processed_data = main_processing_pipeline()
print("Data processing complete!")
processed_data.head()
```
